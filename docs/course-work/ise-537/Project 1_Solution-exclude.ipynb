```python
import pandas as pd
import numpy as np
import math as m
```

```python
raw = pd.read_excel('./data/Project1_TimeSeries_Momentum_Due20200913.xlsx',sheet_name='AssetsPrices',header=1,index_col=0)
raw.head()
```

```python
use_cols = list(raw.columns)[0:15]  #currency: [31:36]
use_cols
```

```python
df=raw[use_cols].sort_index()
df = df[df.index >= pd.to_datetime('2000-12-31')]
df = df[df.index.is_month_end]
df.head()
```

# Assets Monthly returns 

```python
monthly_return = df.pct_change()
monthly_return = monthly_return[monthly_return.index > pd.to_datetime('2001-12-31')]
monthly_return.head()
```

# Asset return for the past 12 months

```python
yearly_return = df.pct_change(periods=12)
yearly_return = yearly_return[yearly_return.index >= pd.to_datetime('2001-12-31')]
yearly_return.head()
```

```python
# Since our momentum strategy make the decision based on the yearly return for the previous month, we build the look up yearly return by shift one month.
# for example: our decision for the monthly return end on 2002-01-31 is based on the yearly return end at 2001-12-31. 
look_up_yearl_return = yearly_return.shift().iloc[1:]
look_up_yearl_return.head()
```

```python
assert monthly_return.shape == look_up_yearl_return.shape
```

# Momentum strategy monthly return

```python
strategy = monthly_return[look_up_yearl_return >= 0]
strategy[look_up_yearl_return < 0] = - monthly_return[look_up_yearl_return < 0]
strategy.head()
```

# Calculate the calender year return 

```python
year = [str(i) for i in range(2002, 2021)]
All_annl_ret = pd.DataFrame()

for i in year:
    All_annl_ret[i] = (strategy.loc[i] + 1).prod() - 1 

All_annl_ret = All_annl_ret.transpose()
All_annl_ret
```

# Calculate the annl. ret from 2002 to current

```python
annl_ret = (strategy + 1).prod() ** (12 / strategy.notnull().sum()) - 1
annl_ret
```

# Calculate the annl. vol over all years

```python
annl_vol = strategy.std() * m.sqrt(12)
annl_vol
```

# Calculate the Sharp ratio

```python
sharpe_ratio = annl_ret / annl_vol
sharpe_ratio
```
